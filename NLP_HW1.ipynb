```python
with open('alice29.txt', 'r') as f :
  text = f.read()
```

## Cleaned Text

```python
# Download stopwords if you haven't done so before
nltk.download('stopwords')

# Load stopwords
stop_words = set(stopwords.words('english'))
```

```python
content = text.lower() #lowercase all alphabet
```

```python
content = re.sub(r'[^\w\s]', '', content) #remove puntuation
content = re.sub(r'\s+', ' ', content).strip() #convert multiple space into single space
```

```python
#Clean stop words
words = content.split()
filtered_words = [word for word in words if word not in stop_words]

cleaned_content = ' '.join(filtered_words)
```

```python
cleaned_content
```

```python
with open('cleaned.txt', 'w', encoding='utf-8') as f:
    f.write(cleaned_content)
```

```python
from textblob import TextBlob
import spacy
import time
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
def nltk_analysis(text):
    start = time.time()
    words = nltk.word_tokenize(text)
    word_frequency = Counter(words)

    pos_tags = nltk.pos_tag(words)
    end = time.time()
    return {"time": end - start, "result": {"total words": len(words), "unique words": len(word_frequency), "part of speech tags": len(pos_tags)}}


def textblob_analysis(text):
    start = time.time()
    blob = TextBlob(text)
    words = blob.words
    word_frequency = Counter(words)

    pos_tags = blob.tags
    end = time.time()
    return {"time": end - start, "result": {"total words": len(words), "unique words": len(word_frequency), "part of speech tags": len(pos_tags)}}


def spacy_analysis(text):
    start = time.time()
    doc = nlp(text)
    words = [token.text for token in doc]
    word_frequency = Counter(words)

    pos_tags = [(token.text, token.pos_) for token in doc]
    end = time.time()
    return {"time": end - start, "result": {"total words": len(words), "unique words": len(word_frequency), "part of speech tags": len(pos_tags)}}
```

```python
nltk_result = nltk_analysis(cleaned_content)
textblob_result = textblob_analysis(cleaned_content)
spacy_result = spacy_analysis(cleaned_content)
```

```python
nltk_result
```

```python
results = (
    f"Performance Comparison:\n"
    f"NLTK: Time: {nltk_result['time']:.4f}s, Total Words: {nltk_result['result']['total words']}, "
    f"Unique Words: {nltk_result['result']['unique words']}, Part Of Speech Tags: {nltk_result['result']['part of speech tags']}\n"
    f"TextBlob: Time: {textblob_result['time']:.4f}s, Words: {textblob_result['result']['total words']}, "
    f"Unique Words: {textblob_result['result']['unique words']}, Part Of Speech Tags: {textblob_result['result']['part of speech tags']}\n"
    f"spaCy: Time: {spacy_result['time']:.4f}s, Words: {spacy_result['result']['total words']}, "
    f"Unique Words: {spacy_result['result']['unique words']}, Part Of Speech Tags: {spacy_result['result']['part of speech tags']}\n"
)
```
